```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('worldometer_data.csv')
data
```

```python
new_data = data[['Continent','Country/Region', 'TotalCases', 'TotalDeaths', 'TotalRecovered', 'ActiveCases', 'Serious,Critical']]
new_data
```

```python
filt1 = new_data['TotalCases'] > 50000
data_filt = new_data[filt1]
data_filt
```

```python
filt1 = new_data['TotalCases'] > 50000
filt2 = new_data['Continent'] == 'Asia'
filt = filt1 & filt2
new_data[filt]
```

```python
data_filt.head()
```

```python
df_replace = data_filt.fillna(0)
df_replace
```

```python
filt3 = df_replace['Continent'] == 'Asia'
filt4 = df_replace['Continent'] == 'Europe'
filt = filt3 | filt4
data_clear = df_replace[filt]
data_clear
```

```python
data_sort = data_clear.sort_values('TotalCases', ascending = True)
data_sort
```

```python
ds_clear = data_sort.set_index('Country/Region')
ds_clear
```

```python
import seaborn as sns
sns.set()

graph = sns.scatterplot(x = 'TotalCases', y = 'TotalRecovered', data = ds_clear)
graph
```

```python
from bokeh.plotting import figure,show
from bokeh.models import HoverTool

TOOLTIPS = HoverTool(tooltips = [
    ('index', '$index'),
    ('(TotalCases, TotalRecovered)', '(@TotalCases, @TotalRecovered)'),
    ('Continent', '@Continent')
])

plot = figure(title ='Data Covid-19', x_axis_label='TotalCases', y_axis_label='TotalRecovered', plot_width=700, plot_height=700, tools=[TOOLTIPS])
plot.circle('TotalCases','TotalRecovered', size=10, source=ds_clear)
show(plot)
```
